```python
#1 Dataframe presidenta female y pais democratico

df_fem_dem = df[(df['is_female_president'] == True) & (df['is_democracy'] == True)]
df_fem_dem.head()
#df_fem_dem.shape
df_fem_dem
```

Para resolver el problema de los años, debo encontrar diferentes soluciones, seguramente habrá alguna mas ingeniosas que otras pero considero que los metodos iterativos para bases de datos no son los mas ingeniosos... Podrian tardarse demasiado y requierir una gran cantidad de computo.

Veo que cada dato por pais está desde 1950 a 2020, pensaba en contar la cantidad de veces que sale el nombre de una de las presidentas, pero eso no podría garantizar que sean periodos continuos de presidencia.


Algo que si noto es que la columna importante es 'election_year' alterna entre NA y el año. cuando hay un cambio entre NA y un valor int es porque hubo un cambio de presidente o una reelección, si o si, entonces, primer paso interesante, retirar los NA (específicos de esa columna), y luego vamos a comparar los años.

```python
df_yr_na = df_fem_dem.dropna(subset='election_year')
#df_yr_na.to_csv('output_file.csv', index=False)
#save

df_yr_na.head()
```

Ya que tenemos los valores unicos, tenemos que comparar los subunicos nombres, es decir, buscamos una clave de nombre especifica (verificar si la columna de nombres es categórica), luego de eso, dado que quitamos los NA, podemos garantizar que entre el ultimo nombre y el 3er nombre

```python
print(df_yr_na['president_name'].dtype)
```

```python
#Cambiamos a categorías
df_yr_na['president_name'] = df_yr_na['president_name'].astype('category')
#print(df_yr_na['president_name'].dtype)
#print(df_yr_na['election_year'].dtype)
#Cambiamos años a int
df_yr_na['election_year'] = df_yr_na['election_year'].astype(int)
print(df_yr_na['election_year'].dtype)
#df_yr_na['president_name'].unique() #Tenemos 30 presidentas unicas.
```

```python
#Vamos a usar un groupby para crear un diccionario de los nombres y años de mandato.

dic_pres = (
    df_yr_na.groupby('president_name', observed=True)['election_year'].agg(list).to_dict()
)

print(dic_pres)
#con agg list, añade cada grupo a una listad
#con to_dict() lo convierte en un diccionario
```

Ahora, organizamos en orden de los años de las listas internas y restamos los limites de esa lista, luego de eso, con eso hacemos el ranking.

```python
for key in dic_pres:
  dic_pres[key].sort(reverse=True)
print(dic_pres)


#Ahora, tomamos el elemento 0 y el elemento ultimo (solo cuando hay mas de un elemento) (quienes tienen mas de un elemento son las que nos interesan)
#y restamos, luego de eso, añadimos al diccionario

for key in dic_pres:
  c=0
  a=0
  b=0
  if len(dic_pres[key]) > 1:
    a = int(dic_pres[key][0])
    b = int(dic_pres[key][-1])
    c = a-b
    #print(c)
    dic_pres[key].append(c)
    #print(dic_pres[key])

print(dic_pres)

```

Listo. Ahora, solo resta comparar el ultimo valor de cada lista y encontrar el mayor.

```python
#print(dic_pres)

mayor = -1
nombre = None
for key in dic_pres:
  if len(dic_pres[key]) > 1:
    period_value = dic_pres[key][-1]
    #print(period_value)

    if period_value > mayor:
      mayor = period_value
      nombre = key
  else:
    continue
print(f'La presidenta con más años continuos de mandato en una democracia fue "{nombre}" con {mayor} años.')
```

```python
#SOLUCIÓN DE CHAT GPT

import pandas as pd

df = pd.read_csv("democracy_data.csv")

# Filtrar presidentas en democracias
df_fem_dem = df[
    (df["is_female_president"]) &
    (df["is_democracy"])
]

# Eliminar filas sin año electoral
df_fem_dem = df_fem_dem.dropna(subset=["election_year"]).copy()

# Convertir tipo
df_fem_dem["election_year"] = df_fem_dem["election_year"].astype(int)

# Agrupar por presidenta y calcular primer y último año
mandates = (
    df_fem_dem
    .groupby("president_name")["election_year"]
    .agg(["min", "max"])
)

# Duración del mandato
mandates["years"] = mandates["max"] - mandates["min"]

# Encontrar la mayor duración
winner = mandates["years"].idxmax()
max_years = mandates["years"].max()

print(f'La presidenta con más años continuos de mandato en una democracia fue "{winner}" con {max_years} años.')
```
